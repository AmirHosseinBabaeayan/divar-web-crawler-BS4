```python
import bs4
import requests
import time
```

```python
def informationGathering(token):
    url = 'https://divar.ir/v/-/{token}'.format(token=token)
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.content)
    informations = soup.find_all("div", class_="kt-group-row-item kt-group-row-item--info-row")
    data = {}
    data['توکن'] = token
    for inf in informations:
        key = inf.find_all("span", class_="kt-group-row-item__title kt-body kt-body--sm")
        value = inf.find_all("span", class_="kt-group-row-item__value")
        data[key[0].text] = value[0].text

    informations = soup.find_all("div", class_="kt-base-row kt-base-row--large kt-unexpandable-row")
    for inf in informations:
        key = inf.find_all("p", class_="kt-base-row__title kt-unexpandable-row__title")
        value = inf.find_all("p", class_="kt-unexpandable-row__value")
        #print(key)
        #print(value)
        if len(key)==0 or len(value) == 0:
            continue
        data[key[0].text] = value[0].text
    return data
```

```python
from tqdm import tqdm
dataSet = []
s = 0
e = 4031
for index, token in tqdm(enumerate(tokens[s:e]), total=len(tokens[s:e]), desc="Processing tokens"):
    #time.sleep(0.01)
    try:
        dataSet.append(informationGathering(token))
    except Exception as e:
        print("An error occurred while gathering information for token at index {}: {}".format(index, str(e)))
```
