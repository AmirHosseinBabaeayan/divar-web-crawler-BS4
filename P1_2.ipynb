```python
df.drop(del_col, axis=1, inplace=True)
df.head()
```

```python
#.kt-description-row__text--primary
import requests
import bs4

def add_img(token):
    url = 'https://divar.ir/v/-/{token}'.format(token=token)
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.content)
    img_tag = soup.find("img", class_="kt-image-block__image")
    if img_tag == None:
        return b''
    image_url = img_tag["src"]
    try:
        image_response = requests.get(image_url)
    except:
        img = b''
    img = image_response.content

    return img


def add_dec(token):
    url = 'https://divar.ir/v/-/{token}'.format(token=token)
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.content)
    informations = soup.find_all("p", class_="kt-description-row__text--primary")
    if len(informations)==0:
        desc = ''
    else:
        desc = informations[0].text
    return desc
```

```python
dfs = df.sample(50)
```

```python
dfs['توضیحات'] = dfs['توکن'].apply(add_dec)
dfs['تصویر'] = dfs['توکن'].apply(add_img)
```
