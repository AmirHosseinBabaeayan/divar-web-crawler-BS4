```python
import pandas as pd
df = pd.read_csv('outputWithoutNan.csv')
df.head()
```

```python
df_filled = df.fillna(df.mode().iloc[0])
```

```python
df_filled.head()
```

```python
df_filled.describe()
```

```python
df_filled.info()
```

```python
farsi_to_english = {
    '۰': '0',
    '۱': '1',
    '۲': '2',
    '۳': '3',
    '۴': '4',
    '۵': '5',
    '۶': '6',
    '۷': '7',
    '۸': '8',
    '۹': '9'
}

for item in df.columns.tolist():
    df_filled[item] = df_filled[item].replace(farsi_to_english, regex=True)
```

```python
df_filled.head()
```

```python
cols = df_filled.columns.tolist()
cols
```

```python
df_filled[cols[-1]] = df_filled[cols[-1]].str.replace(' تومان', '')
df_filled[cols[-3]] = df_filled[cols[-3]].str.replace(' ماه', '')
```

```python
df_filled.head()
```

```python
for i in [1, 2, -1, -3]:
    df_filled[cols[i]] = df_filled[cols[i]].str.replace('٬', '')
    df_filled[cols[i]] = df_filled[cols[i]].str.replace('توافقی', '0').astype(int)
```

```python
df_filled.head()
```

```python
df_filled.info()
```

```python
for column in df_filled.columns:
    new_title = input(f"لطفاً عنوان جدید برای ستون '{column}' را وارد کنید: ")
    df_filled.rename(columns={column: new_title}, inplace=True)
```

```python
df_filled.head()
```

```python
df_org = df_filled.copy()
```

```python
df_filled = df_org.copy()
```

```python
import numpy as np
mean_values = df_filled.mean()
df_filled = df_filled.replace(0, np.nan).fillna(mean_values)
```

```python
df_filled.describe()
```

```python
# Toman to Milion Toman
df_filled.Price = df_filled.Price / (10**6)
```

```python
for column in df_filled.columns:
    nunique_counts = df_filled[column].nunique()
    print("{} : {}".format(column, nunique_counts))
```

```python
# Deleted columns : Token, Fuel
df_filled.drop(['Token', 'Fuel'], axis=1, inplace=True)
df_filled.head()
```

```python
cols = df_filled.columns.to_list()
cols
```

```python
with open('cols.pickle', 'wb') as file:
    pickle.dump(cols, file)
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
Labels = {}
for i in [2, 3, 4, 5, 7]:
    label = {}
    df_filled[cols[i]] = Encoder.fit_transform(df_filled[cols[i]])
    label_names = Encoder.classes_
    for j in range(len(label_names)):
        label[label_names[j]] = j
    Labels[cols[i]] = label

```

```python
print(Labels)
```

```python
import pickle

with open('Labels.pickle', 'wb') as file:
    pickle.dump(Labels, file)
```

```python
df_filled.head(10)
```

```python
X = df_filled.drop(['Price'], axis=1)
y = df_filled['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
scaler_means = Scaler.mean_
Scaler_stds  = Scaler.scale_
```

```python
with open('Scaler_stds.pickle', 'wb') as file:
    pickle.dump(Scaler_stds, file)

with open('scaler_means.pickle', 'wb') as file:
    pickle.dump(scaler_means, file)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R-squared:", r2)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```

```python
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
